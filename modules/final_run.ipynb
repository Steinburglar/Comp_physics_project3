Start with a metropolis algorthm for a 10x10 square lattice:
We will run the metrolpolis algorithm at a variety of temperatures, and plot the equilibrium magnetism as a function of the temperature

```python
T = 1 #arbitrarily initialize T
square_lattice = lattice.Lattice(J = 1)
Squr_alg = metroplois2.Metroplois(square_lattice, T)
Squr_alg.plot_mag_temp(700, 21, 'Metropolis_anim.mp4')
Video('Metropolis_anim.mp4')
```
